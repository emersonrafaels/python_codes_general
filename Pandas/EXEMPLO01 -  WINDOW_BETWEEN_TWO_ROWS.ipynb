```python
def mark_by_duration(dataframe, column_category, 
                     column_time, timedelta_condition, 
                     name_new_column, condition_duration_greater_than=False):
    
    # INICIANDO O SINAL DE CONDIÇÃO
    if condition_duration_greater_than:
        operator_function = operator.gt
    else:
        operator_function = operator.le

    # INICIANDO O DATAFRAME RESULTANTE
    df_result = pd.DataFrame()

    # CRIANDO COLUNA DE COUNT 2MIN
    dataframe[name_new_column] = "-"

    # ORDENANDO POR CATEGORIA E DATAHORA
    dataframe = dataframe.sort_values(by=[column_category, column_time], ascending=True)

    # PERCORRENDO TODAS AS CATEGORIA
    for company in dataframe[column_category].unique():

        # FILTRANDO A CATEGORIA
        df_filter_category = dataframe[dataframe[column_category] == company]
        df_filter_category = df_filter_category.reset_index()

        # PERCORRENDO TODAS AS LINHAS DO DATAFRAME ATUAL
        for idx, row in df_filter_category.iterrows():

            if idx > 0:

                # REALIZANDO OS CALCULOS
                if operator_function((row[column_time] - last_value), datetime.timedelta(minutes=2)):

                    df_filter_category.at[idx, name_new_column] = 1

            # ARMAZENANDO O VALOR DA LINHA ATUAL
            last_value = row[column_time]

        # ARMAZENANDO TODOS OS RESULTADOS PARA A EMPRESA
        df_result = df_result.append(df_filter_category)

    # SALVANDO O RESULTADO FINAL
    df_result = df_result.reset_index()
    
    return df_result
```
